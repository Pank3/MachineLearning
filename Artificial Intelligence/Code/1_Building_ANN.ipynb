```python
dataset = pd.read_csv('Churn_Modelling.csv')
```

```python
X = dataset.iloc[:,3:-1].values
y = dataset.iloc[:,-1].values
```

```python
X
```

## Data Preprocessing

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
X[:,2] = le.fit_transform(X[:,2])
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

ct  = ColumnTransformer(transformers =[
    ('enc', OneHotEncoder(),[1])], remainder ='passthrough')

X = np.array(ct.fit_transform(X))
```

```python
X
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test =train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train  = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

## Building ANN

```python
import tensorflow as tf
```

### Steps : 

1. Initialize the ANN

2. Adding the input layer and the first hiddne layer

3. Adding the second hidden layer

4. Adding Output layer

----
Training

1. Compile teh ANN

2. Training the ANN on the Training Set

------
Make the prediction on single observation

```python
ann = tf.keras.models.Sequential()
```

```python
#Add input layer and hidden layer
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))

#Add hidden layer
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

```python
#Output Layer
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

```python
#Training

ann.compile(optimizer= 'adam',loss='binary_crossentropy', metrics = ['accuracy'])
#catagorical_crossentropy --> For non binayr classification
#binary_crossentropy --> For binary classification
```

```python
ann.fit(x=X_train,y=y_train,batch_size=32, epochs=200)
```

```python
print(ann.predict(sc.transform([[1,0,0,600, 1, 40, 3, 60000, 2, 1, 1, 50000]]))>0.5) 
```

```python
#Predict on test set
y_pred = ann.predict(X_test)
y_pred = y_pred > 0.5
print(np.concatenate((y_pred,np.expand_dims(y_test,axis=1)),axis=1))
# y_pred.shape, y_test.shape
```

## Confusion matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test,y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```
